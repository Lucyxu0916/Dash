```python
import pandas as pd
import dash
import dash_core_components as dcc
import dash_html_components as html
from plotly import __version__
from plotly.offline import init_notebook_mode, plot, iplot
from plotly import graph_objs as go

import dash
import dash_core_components as dcc
import dash_html_components as html
from dash.dependencies import Input,Output

from ipywidgets import interact, interactive, fixed, interact_manual
import ipywidgets as widgets
```

```python
# Read Dataset
US_State = pd.read_csv('https://raw.githubusercontent.com/plotly/datasets/master/2014_usa_states.csv')
US_State["State"]
US_State["Population"]
```

```python
#Create Layout
app=dash.Dash()
app.layout=html.Div([
    ##Title
    html.Div([
        html.H1("2014 US State Population"),
    ]),
    ##Population Histogram
    dcc.Graph(id="histogram"),
    
    ## Dropdown menu to add Animal data
    html.Div([
        dcc.Dropdown(
            id='states',
            options=[{'label': i, 'value': i} for i in list(US_State["State"]) ],
            value= list(US_State["State"])[0:5], 
            multi=True
        )
    ], style={'width': '100%', 'display': 'inline-block'}),
])
# Dropdown menu to add GUID data
@app.callback(
    Output('histogram', 'figure'),
    [Input('state', 'population')]
)
def update_output(x):
    return( histogramData(x) )
```

```python
#Plots
def histogramData(x):
    print(x)
    # dropdown selection returns a list=x
        
    trace= go.Bar(
        x= US_State["State"],
        y= US_State["Population"],
    )
    
    return{
        'data' : [trace],
        'layout': go.Layout(
            xaxis={'title': '<b>State'},
            yaxis={'title': '<b>Population'},
            title='<b>State Population'
        )
    }
```

```python
#Run
if __name__ == '__main__':
    app.run_server(debug=True, host='0.0.0.0', port = 8124)
```
